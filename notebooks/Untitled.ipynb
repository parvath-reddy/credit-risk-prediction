```python
features = [
    "Age",
    "Sex",
    "Job",
    "Housing",
    "Saving accounts",
    "Checking account",
    "Credit amount",
    "Duration"
]
```

```python
target = "Risk"
```

```python
df_model = df[features + [target]].copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_model.drop(target, axis=1) 
```

```python
y = df_model[target] 
```

```python
X
```

```python
target
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=1
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
def train_model(model, param_grid, X_train, y_train, X_test, y_test):
    grid = GridSearchCV(
        model, 
        param_grid, 
        cv=5, 
        scoring="accuracy", 
        n_jobs=-1
    )
    grid.fit(X_train, y_train)
    
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    
    return best_model, acc, grid.best_params_
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize Decision Tree
dt = DecisionTreeClassifier(random_state=1, class_weight="balanced")

# Define hyperparameter grid
dt_param_grid = {
    "max_depth": [3, 5, 7, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
best_dt, acc_dt, params_dt = train_model(dt, dt_param_grid, X_train, y_train, X_test, y_test)


```

```python
print("Decision Tree Accuracy:", acc_dt)
```

```python
print("Best Parameters:", params_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest
rf = RandomForestClassifier(
    random_state=1,
    class_weight="balanced",
    n_jobs=-1
)
```

```python
# Hyperparameter grid
rf_param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [5, 7, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
best_rf, acc_rf, params_rf = train_model(rf, rf_param_grid, X_train, y_train, X_test, y_test)
```

```python
print("random forest accuracy ", acc_rf)
```

```python
print("Best Parameters:", params_rf)
```

```python
et = ExtraTreesClassifier(
    random_state=1,
    class_weight="balanced",
    n_jobs=-1   # use all CPU cores
)
```

```python
et_param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [5, 7, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
best_et, acc_et, params_et = train_model(et, et_param_grid, X_train, y_train, X_test, y_test)
```

```python
print("random forest accuracy ", acc_et)
```

```python
print("Best Parameters:", params_et)
```

```python
# Compute scale_pos_weight for class imbalance
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

# Define XGBoost model
xgb = XGBClassifier(
    random_state=1, 
    scale_pos_weight=scale_pos_weight, 
    eval_metric="logloss"
)

```

```python
# Parameter grid
xgb_param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.01, 0.1, 0.2],
    "subsample": [0.7, 1.0],
    "colsample_bytree": [0.7, 1.0]
}

```

```python
# Train model
best_xgb, acc_xgb, params_xgb = train_model(
    xgb, xgb_param_grid, X_train, y_train, X_test, y_test
)
```

```python
print("XGb accuracy", acc_xgb)
```

```python
print("Best params", params_xgb)
```
